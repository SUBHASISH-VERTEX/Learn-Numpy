# NUMPY EXERSISE

```python
import numpy as np
s = np.array([[5,3,4,6,7,8,9,1,2],          #sum =  45
             [6,7,2,1,9,5,3,4,8],           #sum =  45
             [1,9,8,3,4,2,5,6,7],           #sum =  45
                                                               #in case of rows and coloums line sum is 45
             [8,5,9,7,6,1,4,2,3],           #sum =  45
             [4,2,6,8,5,3,7,9,1],           #sum =  45
             [7,1,3,9,2,4,8,5,6],           #sum =  45

             [9,6,1,5,3,7,2,8,4],           #sum =  45
             [2,8,7,4,1,9,6,3,5],           #sum =  45
             [3,4,5,2,8,6,1,7,9]            #sum =  45
])
```

```python
s.sum()
```

```python
b = np.sum(s,axis = 1)
for i in b:
    if i != 45:
        print("sudo is not valid")
        break
else:
    print("for rows is valid")
```

```python
np.sum(s,axis = 0)
```

```python
s[0:3,0:3]
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        n= s[i:i+3,j:j+3]
        print(n.sum())
```
